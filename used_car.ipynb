```python
col_list = ["mileage_num","engine_num","power_num","Kilometers_Driven"]
for col in col_list:
        df[col] = df[col].fillna(df[col].median())

df.isnull().sum()
```

```python
df = df[df["Price"].notna()].copy()

df.isna().sum()
```

#Data Preparation

```python
x = df.drop(["Price"],axis=1)

y = df["Price"]
```

```python
#One hot encoding
x = pd.get_dummies(
    x,
    columns= x.select_dtypes(include=["object","category"]).columns.to_list(),
    drop_first=True
)

```

```python
x.head()
```

```python
x=x.astype(float)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)

print(f"Train size: {x_train.shape}")
print(f"Train size: {x_test.shape}")
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
# metrics
from sklearn.metrics import root_mean_squared_error, r2_score
y_train_pred = lin_reg.predict(x_train)
y_test_pred = lin_reg.predict(x_test)

print(f"Train rmse: {root_mean_squared_error(y_train,y_train_pred)}")
print(f"Test rmse: {root_mean_squared_error(y_test,y_test_pred)}")
```

```python
# ---------- 1) AFTER BASE MODEL (the one using X_train/X_test) ----------
base_train_rmse = root_mean_squared_error(y_train, y_train_pred)
base_test_rmse  = root_mean_squared_error(y_test,  y_test_pred)
base_train_r2   = r2_score(y_train, y_train_pred)
base_test_r2    = r2_score(y_test,  y_test_pred)
```
